```python
df['new_title'] = ''
```

```python
for i in range(df.shape[0]):
    current_title = df['title'].values[i]
    current_title_list = current_title.split(' ')
    
    for each_word in current_title_list:
        if each_word.lower() in it_specialist_list:
            df['new_title'].values[i] = 'IT Specialist'
            break
        
        elif each_word.lower() in marketing_list:
            df['new_title'].values[i] = 'Marketing'
            break
        
        else:
            df['new_title'].values[i] = 'Others'
```
