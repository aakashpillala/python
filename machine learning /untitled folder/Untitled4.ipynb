```python
from sklearn.datasets import make_blobs
```

```python
X, do_not_use = make_blobs(n_samples=500)
```

```python
X
```

```python
do_not_use
```

```python
from sklearn.cluster import KMeans
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
km = KMeans(n_clusters=4)
```

```python
km
```

```python
km.fit(X)
```

```python
km.cluster_centers_
```
