```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
no_of_skills = np.linspace(1, 40, 1000, dtype='int')
no_of_skills
```

```python
buffer = np.random.randint(1000, 50000, 1000)
buffer
```

```python
salaries = no_of_skills * 8000 + buffer
salaries
```

```python
df = pd.DataFrame({
    'skills' : no_of_skills,
    'salaries' : salaries
})
```

```python
df.to_excel('dataset1.xlsx', index=False)
```

```python
df.head(60)
```

```python
plt.rcParams['figure.figsize'] = 12,10
plt.scatter(no_of_skills, salaries)
plt.xlabel('No of skills')
plt.ylabel('Salaries')
```

```python
X = df['skills'].values
Y = df['salaries'].values
```

```python
X
```

```python
Y
```

```python
X = X.reshape(-1, 1)
X
```

```python
Y = Y.reshape(-1, 1)
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = 0.7, test_size=0.3)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, Y_train)
```

```python
lr.predict([[30]])
```

```python
lr.predict([[5]])
```

```python
lr.predict([[5]])
```

```python
lr.predict([[3]])
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
type(lr)
```

```python
Y_pred = lr.predict(X)
```

```python
plt.rcParams['figure.figsize'] = 12,10
plt.scatter(no_of_skills, salaries)
plt.plot(no_of_skills, Y_pred, color='red')
plt.xlabel('No of skills')
plt.ylabel('Salaries')
```
