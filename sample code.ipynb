```python
import numpy as np 
company = ['google','amzn','msft','fb']
revenue = [90,123,56,43]
profit = [89,45,35,23]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue,label="revenue")
plt.xticks(xpos,company)  #on x if we use ticks it will use names 
plt.ylabel("us technology stocks")
plt.legend(loc='best')
```

```python
plt.bar(xpos-0.2,revenue,width=0.4,label="revenue",color='red')
plt.bar(xpos+0.2,profit,width=0.4,label="profit",color='green')
plt.xticks(xpos,company)  #on x if we use ticks it will use names 
plt.ylabel("revenue")
plt.legend(loc='best') 
plt.title("sample graph")
```

```python
plt.barh(xpos,revenue,label="revenue")#barh is used for horizental bars
plt.xlabel("revenue")  #on x if we use ticks it will use names 
plt.yticks(xpos,company)
plt.legend(loc='best')
```
